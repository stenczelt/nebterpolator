```python
import nebterpolator
from nebterpolator import path_operations

import numpy as np
import ase.io

from nebterpolator.core import bond_connectivity

import networkx as nx
from matplotlib import  pyplot as plt
from matplotlib import gridspec
```

```python
traj = ase.io.read("traj_4.xyz", ":")
len(traj)
```

```python
e0_byhand = {'H': -12.336810713133827,
             'C': -145.60813964301911,
             'O': -429.04802314923108}

def makefig():
    fig = plt.figure()
    fig.set_figheight(8)
    fig.set_figwidth(10)
    return fig, fig.add_subplot()
```

```python
# chemical symbols and colors for the graphs
chemical_symbols = traj[0].get_chemical_symbols()
colors = {"H":"tab:blue", "C":"tab:gray", "O":"tab:red"}
node_colros = [colors[s] for s in chemical_symbols]
```

```python
pos_nm = np.array([at.get_positions() for at in traj]) * 0.1
pos_nm.shape  # so this is (n_frames, n_atoms, 3)
```

# connectivity for the entire trajectory

```python
conn_traj = [bond_connectivity(pos_nm[i], chemical_symbols) for i in range(len(traj))]
graphs_traj = [nx.from_edgelist(ed) for ed in conn_traj]
```

```python
interest = []
changes = []

for i in range(len(traj) - 1):
    this_change = not nx.algorithms.is_isomorphic(graphs_traj[i], graphs_traj[i+1])
    changes.append(this_change)
    if this_change:
        interest.append([i, i+1])

changes = np.array(changes, dtype=int)
print(interest)
```

```python
plt.plot(changes)
```

```python
# fig, ax = makefig()
# nx.draw(graphs_traj[0], with_labels=True, font_weight='bold', node_color=node_colros, ax=ax)
```

```python
fig = plt.figure()
fig.set_figheight(10)
fig.set_figwidth(10)
gs = gridspec.GridSpec(2, 2, figure=fig)

flat_interest = np.array(interest).flat[:]
for i in range(4):
    ax = fig.add_subplot(gs[i])
    nx.draw(graphs_traj[flat_interest[i]], with_labels=True, font_weight='bold', node_color=node_colros, ax=ax)
    ax.set_title("{}, {}".format(i, flat_interest[i]))
```
